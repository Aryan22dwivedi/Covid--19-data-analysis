```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Sample: Read data (replace with your actual CSV path)
df = pd.read_csv(r'C:\Users\LENOVO\Desktop\rise intership\covid -19 data analysis\country_wise_latest (1).csv')
df.head()
```

```python
# Select top 10 countries by confirmed cases
top_countries = df.sort_values(by='Confirmed', ascending=False).head(10)

# Prepare the data for area chart
area_data = top_countries[['Country/Region', 'Active', 'Recovered', 'Deaths']]
area_data.set_index('Country/Region', inplace=True)

# Plot stacked area chart
area_data.plot(kind='area', stacked=True, figsize=(12,6), alpha=0.6)
plt.title('COVID-19 Case Composition in Top 10 Countries')
plt.xlabel('Country')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

Top 10 Countries by Total Confirmed Cases

```python
top10 = df.sort_values(by='Confirmed', ascending=False).head(10)
top10.plot(x='Country/Region', y='Confirmed', kind='barh', title='Top 10 Countries by Confirmed Cases')
```

Top 10 Countries by New Cases 

```python
top10_new = df.sort_values(by='New cases', ascending=False).head(10)
top10_new.plot(x='Country/Region', y='New cases', kind='barh', color='orange', title='Top 10 Countries by New Cases')
```

Top 10 Countries by New Cases 

```python
top10_new = df.sort_values(by='New cases', ascending=False).head(10)
top10_new.plot(x='Country/Region', y='New cases', kind='barh', color='orange', title='Top 10 Countries by New Cases')
```

WHO Region-Wise Summary

```python
region_summary = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum()
region_summary.plot(kind='bar', stacked=True, title='WHO Region-wise COVID-19 Summary')
```
